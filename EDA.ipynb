```python
# -------------------- 🌾 EDA for Crop Yield Dataset --------------------

# Enable inline plotting for Jupyter Notebook
%matplotlib inline  

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# --------------------  Load Dataset --------------------
data_path = os.path.join("data", "crop_yield_predicted.csv")
df = pd.read_csv(data_path)
print("Dataset loaded successfully!\n")
print("Shape:", df.shape)
print("\nColumn info:")
print(df.info())
print("\nMissing values in each column:")
print(df.isnull().sum())
print("\nSummary statistics:")
print(df.describe())

# --------------------  Distribution of Crop Yield --------------------
plt.figure(figsize=(8,5))
sns.histplot(df['Crop_Yield'], bins=30, kde=True, color='green')
plt.title("Distribution of Crop Yield")
plt.xlabel("Crop Yield")
plt.ylabel("Frequency")
plt.show()

# --------------------  Boxplots for Numerical Features --------------------
numerical_cols = ['Temperature', 'Humidity', 'Soil_Quality', 'NPK_Ratio', 'Fertility_Index']
plt.figure(figsize=(12,6))
sns.boxplot(data=df[numerical_cols])
plt.title("Boxplot of Numerical Features")
plt.xticks(rotation=45)
plt.show()

# --------------------  Correlation Heatmap --------------------
plt.figure(figsize=(8,6))
corr = df[numerical_cols + ['Crop_Yield']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# --------------------  Crop Yield by Crop Type --------------------
plt.figure(figsize=(8,5))
sns.boxplot(x='Crop_Type', y='Crop_Yield', data=df)
plt.title("Crop Yield by Crop Type")
plt.show()

# --------------------  Pairplot for Selected Features --------------------
selected_cols = ['Crop_Yield'] + numerical_cols
sns.pairplot(df[selected_cols])
plt.show()

# -------------------- ✅ EDA Completed --------------------
print("Exploratory Data Analysis completed!")
```
