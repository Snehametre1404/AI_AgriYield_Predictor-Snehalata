```python
# Enable inline plotting for Jupyter Notebook
%matplotlib inline  

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("crop_yield_preprocessed.csv")

# Basic dataset information
print("Dataset shape:", df.shape)     # Shape of dataset (rows, columns)
print("\nColumn info:")
print(df.info())                      # Data types and non-null values

# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Summary statistics of numerical columns
print("\nSummary statistics:")
print(df.describe())

# -------------------- Visualization Section --------------------

# Distribution of Crop Yield
plt.figure(figsize=(8,5))
sns.histplot(df['Crop_Yield'], bins=30, kde=True, color='green')
plt.title("Distribution of Crop Yield")
plt.xlabel("Crop Yield")
plt.ylabel("Frequency")
plt.show()

# Boxplots for numerical features to check outliers
numerical_cols = ['Soil_pH', 'Temperature', 'Humidity', 'Wind_Speed', 
                  'N', 'P', 'K', 'Soil_Quality', 'NPK_Ratio', 'Fertility_Index']
plt.figure(figsize=(15,7))
sns.boxplot(data=df[numerical_cols])
plt.title("Boxplot of Numerical Features")
plt.xticks(rotation=45)
plt.show()

# Correlation heatmap to see relationships between features
plt.figure(figsize=(12,8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# Crop Yield distribution by Crop Type
plt.figure(figsize=(8,5))
sns.boxplot(x='Crop_Type', y='Crop_Yield', data=df)
plt.title("Crop Yield by Crop Type")
plt.show()

# Crop Yield distribution by Soil Type
plt.figure(figsize=(8,5))
sns.boxplot(x='Soil_Type', y='Crop_Yield', data=df)
plt.title("Crop Yield by Soil Type")
plt.show()

# Crop Yield distribution by Season
plt.figure(figsize=(8,5))
sns.boxplot(x='Season', y='Crop_Yield', data=df)
plt.title("Crop Yield by Season (1=Rabi, 2=Summer, 3=Kharif)")
plt.show()

# Pairplot for selected important features
selected_cols = ['Crop_Yield', 'N', 'P', 'K', 'Soil_Quality', 'Temperature', 'Humidity']
sns.pairplot(df[selected_cols])
plt.show()

# Final message
print("EDA completed!")  # Comment line: End of Exploratory Data Analysis
```
